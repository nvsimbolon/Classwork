### a.	Use the column ‘geographic_location’ to create a new column called ‘continent’. Merge the values of the ‘geographic column appropriately to obtain 6 distinct values for the ‘continent’ column – Asia, Africa, North America, South America, Europe and Oceania. Drop the column ‘geographic_location’ (2 points for code).

```python
# Creating the continent column
def area (row):
    if row['geographic_location'] == 'Western Asia' :
      return 'Asia'
    if row['geographic_location'] == 'South-eastern Asia' :
      return 'Asia'
    if row['geographic_location'] == 'Southern Asia' :
      return 'Asia'
    if row['geographic_location'] == 'Eastern Asia' :
      return 'Asia'
    if row['geographic_location'] == 'Central Asia' :
      return 'Asia'
    if row['geographic_location'] == 'Northern Africa' :
      return 'Africa'
    if row['geographic_location'] == 'Western Africa' :
      return 'Africa'
    if row['geographic_location'] == 'Middle Africa' :
      return 'Africa'
    if row['geographic_location'] == 'Southern Africa' :
      return 'Africa'
    if row['geographic_location'] == 'Eastern Africa' :
      return 'Africa'
    if row['geographic_location'] == 'North America' :
      return 'North America'
    if row['geographic_location'] == 'Northern America' :
      return 'North America'
    if row['geographic_location'] == 'South America' :
      return 'South America'
    if row['geographic_location'] == 'Northern Europe' :
      return 'Europe'
    if row['geographic_location'] == 'Southern Europe' :
      return 'Europe'
    if row['geographic_location'] == 'Eastern Europe' :
      return 'Europe'
    if row['geographic_location'] == 'Western Europe' :
      return 'Europe'
    return 'Oceania'

social['Continent'] = social.apply(area, axis=1)

# Dropping the geographic_location column
social.drop(columns = 'geographic_location', inplace = True)
```

### b.	We need to find the African country that is the closest to country G with regard to social indicators. Perform the following steps:
#### i.	Standardize the social indicator columns appropriately so that all of them are on the same scale (remember to exclude GDP per capita from social indicators). (2 points for code)


```python
# Creating standardizing function
f = lambda x: (x-x.min())/(x.max() - x.min())

# Applying f function
soc_ind = social.iloc[:,1:9].apply(f)

# Dropping the gdpPerCapita column
soc_ind.drop(columns = 'gdpPerCapita', inplace = True)
soc_ind
```

#### ii.	Compute the Manhattan distance between country G and each of the African countries. (2 points for code)

```python
ind = social.index[social['country'] == 'Luxembourg']
africa = social[social['Continent'] == 'Africa']

africa_ind = africa.iloc[:,1:9].apply(f)
africa_ind.drop(columns = 'gdpPerCapita', inplace = True)

africaminusl = africa_ind - soc_ind.loc[ind[0],:]
africaminusl = africaminusl.pow(2)
distances = africaminusl.sum(axis=1)
distances_sorted = distances.sort_values()
```
